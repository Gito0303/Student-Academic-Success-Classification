### Deskripsi Data

Klasifikasi dilakukan berdasarkan beberapa faktor, termasuk:

>   - **Faktor Demografis** : Status pernikahan, jenis kelamin, usia saat pendaftaran.
>   - **Faktor Pendidikan** : Kualifikasi pendidikan sebelumnya beserta nilainya, pekerjaan dan kualifikasi orang tua, kebutuhan pendidikan khusus.
>   - **Faktor Geografis** : Apakah mahasiswa berasal dari luar daerah atau internasional.
>   - **Faktor Keuangan** : Status beasiswa dan hutang.
>   - **Faktor Akademik** : Jumlah mata kuliah yang diambil, dievaluasi, dan disetujui pada semester 1 dan 2.
>   - **Faktor Kehadiran** : Mode kehadiran (siang atau malam).

---

**Target klasifikasi** merujuk pada status akhir akademik mahasiswa, yang terbagi dalam tiga kategori utama:

1. **Dropout** : Mahasiswa yang berhenti kuliah sebelum menyelesaikan program studinya.
2. **Enrolled** : Mahasiswa yang masih terdaftar aktif dan sedang melanjutkan proses perkuliahan.
3. **Graduate** : Mahasiswa yang menyelesaikan program studi dan lulus.

---

Klasifikasi ini bertujuan untuk memahami dan memprediksi status akademik mahasiswa berdasarkan data historis yang tersedia.

### Tujuan Proyek

1. Mengembangkan model klasifikasi yang dapat memprediksi status akademik mahasiswa (dropout, enrolled, graduate) berdasarkan faktor-faktor yang tersedia.
2. Mengevaluasi performa model deep learning dan machine learning (FNN, DNN dan Random Forest) untuk menangani tugas klasifikasi ini.
3. Program ini akan menghasilkan evaluasi dan hasil pelatihan model yang akan digunakan untuk membangun sebuah sistem klasifikasi sederhana menggunakan Streamlit.

# **PREPROCESSING & TRANSFORMASI**

```python
label_encoder = LabelEncoder()
data['Target'] = label_encoder.fit_transform(data['Target'])
data.sample(3)
```

```python
scaler = MinMaxScaler()

normalisasi_kolom = [
    'Previous qualification (grade)', 'Age at enrollment',
    'Curricular units 1st sem (enrolled)', 'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)', 'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (enrolled)', 'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)', 'Curricular units 2nd sem (without evaluations)'
]

data[normalisasi_kolom] = scaler.fit_transform(data[normalisasi_kolom])
data.sample(3)
```

```python
joblib.dump(scaler, 'scaler.pkl')
```

```python
X_data = data.drop('Target', axis=1)
y_data = data['Target']
```

```python
X_train_dl, X_test_dl, y_train_dl, y_test_dl = train_test_split(X_data, y_data, test_size=0.2, random_state=42, stratify=y_data)
X_train_dl.shape, X_test_dl.shape, y_train_dl.shape, y_test_dl.shape
```

# **PEMODELAN**

## Feedforward Neural Network (FNN)

***

> **Feedforward Neural Network (FNN)** : *Model jaringan saraf sederhana di mana data bergerak satu arah dari input ke output tanpa loop, digunakan untuk tugas klasifikasi atau regresi.*

***

```python
input_shape_dl = (X_train_dl.shape[1],)
dl_model = Sequential([
    Dense(128, activation='relu', input_shape=input_shape_dl),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(len(label_encoder.classes_), activation='softmax')
])

dl_model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history_dl = dl_model.fit(X_train_dl, y_train_dl, epochs=30, batch_size=64, validation_split=0.2, verbose=1)
```

```python
y_pred_dl = dl_model.predict(X_test_dl).argmax(axis=1)
```

## Deep Neural Network (DNN)

***

> **Deep Neural Network (DNN)** : *Jaringan saraf dengan beberapa lapisan tersembunyi yang memungkinkan model menangkap pola yang lebih kompleks dalam data.*

***

```python
input_shape_dnn = (X_train_dl.shape[1],)
dnn_model = Sequential([
    Dense(512, activation='relu', input_shape=input_shape_dnn),
    Dropout(0.2),
    Dense(256, activation='relu'),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(len(label_encoder.classes_), activation='softmax')
])

dnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history_dnn = dnn_model.fit(X_train_dl, y_train_dl, epochs=30, batch_size=64, validation_split=0.2, verbose=1)
```

```python
y_pred_dnn = dnn_model.predict(X_test_dl).argmax(axis=1)
```

## Random Forest

***

> **Random Forest** : *Algoritma ensemble yang menggunakan banyak pohon keputusan untuk meningkatkan akurasi dengan menggabungkan prediksi dari banyak model pohon yang dilatih pada data yang berbeda.*

***

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_dl, y_train_dl)
```

```python
y_pred_rf = rf_model.predict(X_test_dl)
```

# **EVALUASI**

## Feedforward Neural Network (FNN)



> Learning Curves



```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(history_dl.history['loss'], label='Training Loss', color='blue')
axes[0].plot(history_dl.history['val_loss'], label='Validation Loss', color='orange')
axes[0].set_title('Loss Curve - FNN')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Loss')
axes[0].legend()

axes[1].plot(history_dl.history['accuracy'], label='Training Accuracy', color='blue')
axes[1].plot(history_dl.history['val_accuracy'], label='Validation Accuracy', color='orange')
axes[1].set_title('Accuracy Curve - FNN')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy')
axes[1].legend()

plt.tight_layout()
plt.show()
```



> Confusion Matrix



```python
cm_dl = confusion_matrix(y_test_dl, y_pred_dl)
disp_dl = ConfusionMatrixDisplay(confusion_matrix=cm_dl, display_labels=label_encoder.classes_)
disp_dl.plot(cmap='Blues')
```



> Classification Report



```python
report_dl = classification_report(y_test_dl, y_pred_dl, target_names=label_encoder.classes_)
print(report_dl)
```
